```python
def student_graph(roll):
    student = df[df["RollNo"] == roll]
    name = student["Name"].values[0]
    marks = student[subjects].values.flatten()
    
    plt.figure()
    plt.bar(subjects, marks)
    plt.title(f"Subject-wise Marks of {name}")
    plt.xlabel("Subjects")
    plt.ylabel("Marks")
    plt.show()

student_graph(11)   #change roll no here!!
```

```python
# Academic score out of 100
df["Academic_Score"] = (df["Total_Marks"] / 500) * 100

# Performance Index calculation
df["Performance_Index"] = (0.7 * df["Academic_Score"]) + (0.3 * df["Activity_Marks"])

# Verify column creation
df[["RollNo", "Name", "Performance_Index"]].head()
```

```python
import matplotlib.pyplot as plt

class_avg = df["Performance_Index"].mean()

plt.figure()
plt.plot(df["RollNo"], df["Performance_Index"], marker='o', label="Students")
plt.axhline(class_avg, linestyle='--', label="Class Average")
plt.xlabel("Roll Number")
plt.ylabel("Performance Index")
plt.title("Performance Index Trend of Students")
plt.legend()
plt.grid(True)
plt.show()
```
